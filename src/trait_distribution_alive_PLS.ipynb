```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

table = pd.read_csv("/home/bianca/bianca/CAETE-DVM-alloc-allom/src/PLS_alive_traits_MAN_30prec_1y.csv")
table

# Calcula a média ponderada do SLA para cada ano
weighted_mean_g1_per_year = (table.groupby(['YEAR'])['g1']
                               .apply(lambda x: (x * table.loc[x.index, 'OC']).sum() / table.loc[x.index, 'OC'].sum())
                               .reset_index())

# Calcula a variância do SLA para cada ano
variance_g1_per_year = (table.groupby(['YEAR'])['g1']
                         .var()
                         .reset_index()
                         .rename(columns={'g1': 'Variance'}))

# Mescla os resultados da média ponderada e da variância
result = pd.merge(weighted_mean_g1_per_year, variance_g1_per_year, on='YEAR')
result
# Cria o gráfico de linha com a média ponderada do SLA e intervalo de confiança
plt.figure(figsize=(10, 6))
sns.lineplot(data=result, x='YEAR', y='g1', label='Média Ponderada do g1')
# plt.fill_between(result['YEAR'], result['g1'] - result['Variance'], result['g1'] + result['Variance'], alpha=0.3)

plt.title('')
plt.xlabel('Year')
plt.ylabel('WD weighted mean (g/cm3)')
plt.legend()
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcula a média ponderada do SLA para cada ano
weighted_mean_wd_random_per_year = (table.groupby(['YEAR'])['wd_random']
                               .apply(lambda x: (x * table.loc[x.index, 'OC']).sum() / table.loc[x.index, 'OC'].sum())
                               .reset_index())

# Calcula a variância do SLA para cada ano
variance_wd_random_per_year = (table.groupby(['YEAR'])['wd_random']
                         .var()
                         .reset_index()
                         .rename(columns={'wd_random': 'Variance'}))

# Mescla os resultados da média ponderada e da variância
result = pd.merge(weighted_mean_wd_random_per_year, variance_wd_random_per_year, on='YEAR')
result
# Cria o gráfico de linha com a média ponderada do SLA e intervalo de confiança
plt.figure(figsize=(10, 6))
sns.lineplot(data=result, x='YEAR', y='wd_random', label='Média Ponderada do wd_random')
plt.fill_between(result['YEAR'], result['wd_random'] - result['Variance'], result['wd_random'] + result['Variance'], alpha=0.3)

plt.title('')
plt.xlabel('Year')
plt.ylabel('WD weighted mean (g/cm3)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcula a média ponderada do SLA para cada ano
weighted_mean_sla_per_year = (table.groupby(['YEAR'])['wd_random']
                               .apply(lambda x: (x * table.loc[x.index, 'OC']).sum() / table.loc[x.index, 'OC'].sum())
                               .reset_index())

# Calcula a variância do SLA para cada ano
variance_sla_per_year = (table.groupby(['YEAR'])['wd_random']
                         .var()
                         .reset_index()
                         .rename(columns={'wd_random': 'Variance'}))

# Mescla os resultados da média ponderada e da variância
result = pd.merge(weighted_mean_sla_per_year, variance_sla_per_year, on='YEAR')

# Cria o gráfico de linha com a média ponderada do SLA e barras de erro representando a variância
plt.figure(figsize=(10, 6))
sns.lineplot(data=result, x='YEAR', y='wd_random', marker='o', label='Média Ponderada do SLA')
plt.errorbar(result['YEAR'], result['wd_random'], yerr=result['Variance'], fmt='o', capsize=5, label='Variabilidade')

plt.title('Mudança na Média Ponderada do SLA da Comunidade ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Média Ponderada do SLA')
plt.legend()
plt.show()
```
